```python
df_csv = pd.read_csv("ventas_tienda_festival.csv")

# Asi se crea una nueva columna dentro del archivo y multiplicamos precio * unidades

df_csv["importe total"] = df_csv["precio_eur"] * df_csv["unidades"]  

# Con este comando filtramos por importe y se utiliza . copy para no modificar la tabla.
filtrado = df_csv[df_csv["importe total"] >= 30].copy() 

# Con to_html() creamos este csv a html con los caambios, 
# con index=False quitamos los indices, pero sin cabecera de html.

#filtrado.to_html("nueva_tabla.html", index=False)     #Comentado para ver como funciona el siguiente paso

# Con pasamos este csv a html con cabecera
filtrado.to_html('mi_nuevo_archivo.html', index=False)

mi_tabla = filtrado.to_html(index=False)

croquis_html = f""" 
<!DOCTYPE html>
<html>
<head>
<meta charset="UTF-8">
<title>Ventas filtradas</title>
</head>
<body>
<h1>Ventas filtradas (importe ≥ 30 €)</h1>
{mi_tabla}
</body>
</html>
"""

with open("mi_archivo.html", "w", encoding="utf-8") as f:
    f.write(croquis_html)

display(filtrado)
display(df_csv)   
```
